```python
simulated_mutations_counts = pd.read_csv('simulated_mutations_counts.csv', index_col=0)
```

```python
experimental_mutations_counts = pd.read_csv('experimental_mutations_counts.csv', index_col=0)
```

```python
all_muts = pd.DataFrame(index=experimental_mutations_counts.index)
all_muts['exp'] = experimental_mutations_counts
all_muts['sim'] = simulated_mutations_counts.sum(axis=1)/25 #axis=1 sums across rows
```

```python
hist(all_muts.exp, alpha=0.5, label='experimental', bins=10)
title("experimental and simulated mutations distribution")
xlabel("number of lines with a single gene mutated")
ylabel("number of genes with the gene mutated in x lines")
hist(all_muts.sim, alpha=0.5, label='simulated', bins=10)
legend()
savefig('experimental_simulated_histogram.png')
```
